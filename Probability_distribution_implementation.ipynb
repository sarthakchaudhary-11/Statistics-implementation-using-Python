```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#Normal distribution >> Gaussian distribution >> Majority of data points are clustered around mean.

mu = 0
sigma=0.1
s = np.random.normal(mu, sigma,size=1000)
```

```python
plt.hist(s, 30)
plt.show()
```

```python
sns.distplot(s)
```

```python
#Uniform distribution >> probability distribution where every possible outcome has an equal likelihood of happening.

u = np.random.uniform(0,1,size=100)
plt.hist(u)
plt.show()
```

```python
sns.distplot(u)
```

```python
#Log normal distribution
l = np.random.lognormal(mean=0, sigma=0.5, size=1000)
plt.hist(l,bins=100)
plt.show()
```

```python
sns.boxplot(l) #Lognormal is right skewed so it has lot of outliers.
```

```python
sns.distplot(l)
```

```python
 #square root transformation
 sns.distplot(np.sqrt(l))
```

```python
#Chi-Square Distributionc
c = np.random.chisquare(df=10,size=10000)
plt.hist(c,bins=1000)
plt.show()
```

```python
#Multiple chi_square distribution in one plot
c = np.random.chisquare(df=10 , size= 10000)
d = np.random.chisquare(df=20 , size= 10000)
plt.hist(c , bins = 1000)
plt.hist(d , bins = 1000)
plt.show()
```

```python
#t-distribution
t = np.random.standard_t(df = 10, size = 10000)
plt.hist(t , bins = 1000)
plt.show()
```

```python
#F-distribution
f = np.random.f(dfnum = 100 , dfden = 100 , size = 10000)
plt.hist(f , bins = 1000)
plt.show()
```

```python
#discrete distribution
#bernoulli distribution >> binomial distribution with one trial is bernoulli distribution
b = np.random.binomial(n=1 , p=0.5 , size = 10000)
plt.hist(b)
plt.show()
```

```python
#binomial distribution
b = np.random.binomial(n=100 , p=0.5 , size = 10000)
plt.hist(b)
plt.show()
```

```python
#poisson distribution
p = np.random.poisson(lam = 10 , size = 10000)
plt.hist(p , bins = 20)
plt.show()
```
